```python
import datetime

import matplotlib.pyplot as plt
from esm import Alphabet, pretrained
import torch
from tqdm import tqdm
import pandas as pd
import numpy as np
import scipy.stats as ss

import requests as r
from Bio import SeqIO
from io import StringIO
    
```

```python
# # Load sequence to run DMS study over
# cID = 'P0DTC2'

# baseUrl = "http://www.uniprot.org/uniprot/"
# currentUrl = baseUrl+cID+".fasta"
# response = r.post(currentUrl)
# cData = ''.join(response.text)

# Seq=StringIO(cData)
# pSeq=SeqIO.parse(Seq,'fasta')

# # Truncate seqence so that ESM model accepts it: 
# sequence = pSeq[0].seq[0:1024]
# data = [("Sars_Cov_2", sequence)]

# Just the RBD part of the sequence
# sequence = 'RVQPTESIVRFPNITNLCPFGEVFNATRFASVYAWNRKRISNCVADYSVLYNSASFSTFKCYGVSPTKLNDLCFTNVYADSFVIRGDEVRQIAPGQTGKIADYNYKLPDDFTGCVIAWNSNNLDSKVGGNYNYLYRLFRKSNLKPFERDISTEIYQAGSTPCNGVEGFNCYFPLQSYGFQPTNGVGYQPYRVVVLSFELLHAPATVCGPKKSTNLVKNKCVNF' # From the UniProt website
sequence = 'NITNLCPFGEVFNATRFASVYAWNRKRISNCVADYSVLYNSASFSTFKCYGVSPTKLNDLCFTNVYADSFVIRGDEVRQIAPGQTGKIADYNYKLPDDFTGCVIAWNSNNLDSKVGGNYNYLYRLFRKSNLKPFERDISTEIYQAGSTPCNGVEGFNCYFPLQSYGFQPTNGVGYQPYRVVVLSFELLHAPATVCGPKKST' # from the Dataframe
data = [("Sars_Cov_2", sequence)]
```

```python
# Make list of models to run DMS with
models = ['esm1v_t33_650M_UR90S_'+str(i) for i in range(1,6) ] 
#models = ['esm1v_t33_650M_UR90S_1', 'esm1v_t33_650M_UR90S_2', 'esm1v_t33_650M_UR90S_3', 'esm1v_t33_650M_UR90S_4', 'esm1v_t33_650M_UR90S_5']
```

```python
# Import csv file containing relvant mutaitons
url = "https://media.githubusercontent.com/media/jbloomlab/SARS-CoV-2-RBD_DMS/master/results/single_mut_effects/single_mut_effects.csv"
cov_df = pd.read_csv(url)
real_seq = {}
for i,aa in cov_df.iterrows():
    real_seq[aa["site_RBD"]] = aa["wildtype"]
print("".join(real_seq.values()))
```

```python
# # Reorder cov_df in terms of effect and absolute position, remove extraneous index column
# cov_df = cov_df.sort_values(['site_SARS2','mutation'])
# cov_df = cov_df.reset_index()
# cov_df.drop("index",axis=1)

# Reorder cov_df in terms of effect and RBD position, remove extraneous index column
cov_df = cov_df.sort_values(['site_RBD','mutation'])
cov_df = cov_df.reset_index()
cov_df.drop("index",axis=1)
```

```python
# Sanity check cov_df by printing it
cov_df
```

```python
# Model eval activates/deactivates the relevant layers for running the network. See this stack exchange answer for a
# slightly more coherent description: https://stackoverflow.com/questions/60018578/what-does-model-eval-do-in-pytorch
torch.hub.set_dir("../T4Lys/models") # modify location for different machines

# Loop over the models
for mod in models: 
    
    model, alphabet = pretrained.load_model_and_alphabet(mod)
    model.eval()
    if torch.cuda.is_available():
        model = model.cuda()
        print("Transferred model to GPU")
    
    # Load batch converters
    batch_converter = alphabet.get_batch_converter()
    batch_labels, batch_strs, batch_tokens = batch_converter(data)
    
    # Conduct DMS study for the given model
    dms = pd.DataFrame()
    for i in tqdm(range(batch_tokens.size(1))):
        batch_tokens_masked = batch_tokens.clone()
        batch_tokens_masked[0, i] = alphabet.mask_idx
        with torch.no_grad():
            token_probs = torch.log_softmax(
                #model(batch_tokens_masked)["logits"], dim=-1 # if gpu not available, use this
                model(batch_tokens_masked.cuda())["logits"], dim=-1
            )
        col = pd.DataFrame(token_probs[:, i].cpu())
        col = col.transpose()
        dms = pd.concat([dms,col],axis=1)
        
    # Relabel indices
    idx_map = {0: '<cls>', 1: '<pad>', 2: '<eos>', 3: '<unk>', 4: 'L', 5: 'A', 6: 'G', 7: 'V', 8: 'S', 9: 'E', 10: 'R', 11: 'T', 12: 'I', 13: 'D', 14: 'P', 15: 'K', 16: 'Q', 17: 'N', 18: 'F', 19: 'Y', 20: 'M', 21: 'H', 22: 'W', 23: 'C', 24: 'X', 25: 'B', 26: 'U', 27: 'Z', 28: 'O', 29: '.', 30: '-', 31: '<null_1>', 32: '<mask>'}
    dms = dms.rename(index = idx_map)
    
    # Relabel Columns
    new_col = list(sequence)
    new_col.append('<eos>')
    new_col.insert(0,'<cls>')
    dms.columns = new_col
    
    # Delete extraneous predictions and rare amino acids from DMS dataframe. Delete eos and cls token form sequence
    dms = dms.drop(index = ['<cls>','<pad>','<unk>','.','-','<null_1>','<mask>','X','B','U','Z','O'])
    dms = dms.drop(columns = ['<cls>','<eos>']) # DMS now has dimensions L x 20+1
    
    # Save DMS matrix
    dms.to_csv("dms_cov_esm_"+str(models.index(mod))+"_on_"+str(datetime.datetime.now()).split(".")[0].replace(":","_").replace(" ","_at_")+".csv")
    
    # Isolate the Pseudoenergies of the relevant mutations
    cov_df["logP_mut_"+mod] = ""
    cov_df["logP_wt_"+mod] = ""
    cov_df["log_mut_minus_wt_vals_"+mod] = ""
    for i, row in cov_df.iterrows():
        #wt, idx, mt = row["wildtype"], int(row["site_SARS2"]), row["mutant"] # Absolute index
        wt, idx, mt = row["wildtype"], int(row["site_RBD"]), row["mutant"] # Domain index
        assert dms.columns.values.tolist()[idx-1] == wt,  "Wildtype at index "+str(i)+" does not match provided mutation"
        mut_pseudo_e = float(pd.DataFrame(dms.iloc[:, idx]).loc[mt if mt != '*' else '<eos>'])
        wt_pseudo_e = float(pd.DataFrame(dms.iloc[:, idx]).loc[wt])
        cov_df.loc[i,"logP_mut_"+mod] = mut_pseudo_e 
        cov_df.loc[i,"logP_wt_"+mod]  = wt_pseudo_e
        cov_df.loc[i,"log_mut_minus_wt_vals_"+mod] = float(mut_pseudo_e - wt_pseudo_e)
    
    print("Finished running "+mod)
```

```python
# ----------------------------------------------- Plot probabilities as a vector -------------------------------------------------

# Create 3 figures to hold up to two models each
for i in range(1,4):
    # Create figure and relevant axes
    #fig,ax = plt.subplots(1,2,figsize=(10,10),dpi=300,gridspec_kw={'width_ratios': [4, 1,4,1,4, 1,4,1]}) # old method for creating axis
    fig,ax = plt.subplots(1,4,figsize=(12,10),dpi=300)
    fig.tight_layout(h_pad=5)

    # # Divide figure by esm model
    # line = plt.Line2D((.5,.5),(.05,.95), color="k", linewidth=2)
    # fig.add_artist(line)
    
    # create variable to select correct axis
    ax_count = 0
    
    # iterate over two models at a time:
    for j in range(2*i-1,2*i+1):
        # Make sure loop doesn't reach sixth model
        if j == 6:
            ax[2].axis("off")
            ax[3].axis("off")
            break
        
        # save model name
        model = "esm1v_t33_650M_UR90S_"+str(j)
        
        # Get logP_mut - logP_wt data
        log_mut_minus_wt_val = np.array(t4_df["log_mut_minus_wt_vals_"+model]).astype(float).reshape(40,1) # Note that imshow expects data
        
        # Plot log(P_mt)-log(P_wt) at each site 
        im1 = ax[ax_count].imshow(log_mut_minus_wt_val,cmap='RdBu')
        ax[ax_count].set_title(r'$\logP_{mut}-\logP_{wt}$ for esm '+str(j),loc = "right")
        ax[ax_count].set_xticks([])
        ax[ax_count].set_yticks(range(len(t4_df['variant'])))
        ax[ax_count].set_yticklabels(labels=zip(t4_df['variant'].to_numpy(),t4_df['effect'].to_numpy()))
        fig.colorbar(im1, ax=ax[ax_count],orientation='vertical', aspect=20, pad=0.1,shrink = .25)

        # Get logP data
        mut_array = np.array(t4_df["logP_mut_"+model]).astype(float).reshape(40,1)
        
        # Plot just log(P_mt)
        im1 = ax[ax_count+1].imshow(mut_array,cmap='RdBu')
        ax[ax_count+1].set_title(r'$\logP_{mut}$ for esm '+str(j),loc = "right")
        ax[ax_count+1].set_xticks([])
        ax[ax_count+1].set_yticks(range(len(t4_df['variant'])))
        ax[ax_count+1].set_yticklabels(labels=zip(t4_df['variant'].to_numpy(),t4_df['effect'].to_numpy()))
        fig.colorbar(im1, ax=ax[ax_count+1],orientation='vertical', aspect=20, pad=0.1,shrink = .25)

        # iterate axis count
        ax_count += 2
        
    # Save figure, and sanity check self
    plt.savefig("Prob_Vectors_esm_"+str(2*i-1)+"_and_"+str(2*i)+".png",format="png")
    print("Completed Prob_Vectors_esm_"+str(2*i-1)+"_and_"+str(2*i))
```

```python
# ---------------------------------------- Compute Correlations ----------------------------------------------
# Isolate DeltaG
deltaG = np.array(t4_df["ddG"])
deltaG = deltaG.astype(float)
        
deltaG
```

```python
# Remove NaN values from deltaG
nan_values = ~np.isnan(deltaG)
deltaG_cor = deltaG[nan_values].flatten()
```

```python
# Iterate over model columns and calculate correlation
with open("correllations.txt",'w') as f:
    for model in models: 
        # Remove Nan values from all three arrays of interest
        mut_array_cor = np.array(t4_df["logP_mut_"+model])[nan_values].flatten()
        log_mut_minus_wt_cor = np.array(t4_df["log_mut_minus_wt_vals_"+model])[nan_values].flatten()

        # Calculate correlations between deltaG values, logP_mut, and logP_mut-logP_wt
        mut_min_wt_ddG_cor,p_mut_min_wt = ss.pearsonr(log_mut_minus_wt_cor,deltaG_cor)
        mut_array_ddG_cor,p_mut = ss.pearsonr(mut_array_cor,deltaG_cor)

        # Create text describing correlations
        text = '\n'.join(("For the "+model+" model:",
                         "\t logP_mut-logP_wt correlation "+ str( round(mut_min_wt_ddG_cor,3)) + " with a p-value of " + str(round(p_mut_min_wt,3)), 
                         "\t logP_mut correlation "+ str(round(mut_array_ddG_cor,3)) + " a with p-value of " + str(round(p_mut_min_wt,3)),
                         "\n"))
        
        # Print correlations text
        print(text)
        
        # and write it to a file
        f.write(text)
```

```python
# ----------------------- Plot pseudoenergies on line for normalized esm predictions( i.e logP_mut-logP_wt) ---------------------------------
# Note that to plot the unnormalized values (i.e just logP_mut), all you have to do is change the column from which you get the "destable" 
# "neutral" arrays. 

# Make figure
fig,ax = plt.subplots(2,3,figsize=(10,10),dpi=300)

# Iterate over models
for i in range(len(models)): 
    
    # Set up plot
    model = "esm1v_t33_650M_UR90S_"+str(i+1)
    ax.flat[i].axhline(1,-4,4)  # Draw a horizontal line
    ax.flat[i].set_xlim(-4,4)
    ax.flat[i].set_title(model)
    
    # Get destabilizing and neutral events
    destable  = np.array(t4_df.loc[t4_df['effect'] == 'Destabilizing',"log_mut_minus_wt_vals_"+model])
    neutral  = np.array(t4_df.loc[t4_df['effect'] == 'Neutral',"log_mut_minus_wt_vals_"+model])
    
    # Plot destabilizing and neutral events
    ax.flat[i].eventplot(destable, orientation='horizontal', colors='r',label="destabilizing")
    ax.flat[i].eventplot(neutral, orientation='horizontal', linestyles = 'dashed', colors='b',label="neutral")
    ax.flat[i].legend()
    ax.flat[i].axis("off")
    
    # Sanity check
    print("Finish "+model)

ax.flat[5].axis("off")
plt.savefig("Pseudoenergies_on_a_line_mut_minus_wt.png",format='png')
plt.show()

```

```python
# --------------------Plot pseudoenergies on line for unnormalized esm predictions( i.e just logP_mut) ---------------------------------

# Make figure
fig,ax = plt.subplots(2,3,figsize=(10,10),dpi=300)

# Iterate over models
for i in range(len(models)): 
    
    # Set up plot
    model = "esm1v_t33_650M_UR90S_"+str(i+1)
    ax[i].hlines(1,-4,4)  # Draw a horizontal line
    ax[i].xlim(-4,4)
    ax[i].set_title(model)
    
    # Get destabilizing and neutral events
    destable  = np.array(t4_df.loc[t4_df['effect'] == 'Destabilizing',"log_mut_minus_wt_vals_"+model])
    neutral  = np.array(t4_df.loc[t4_df['effect'] == 'Neutral',"log_mut_minus_wt_vals_"+model])
    
    # Plot destabilizing and neutral events
    ax[i].eventplot(destable, orientation='horizontal', colors='r')
    ax[i].eventplot(neutral, orientation='horizontal', linestyles = 'dashed', colors='b')
    ax[i].axis("off")
    
    # Sanity check
    print("Finish "+model)

plt.show()
```

```python
# --------------------------------------------------------- Plot ROC curves --------------------------------------------------------

# Get number of destabilizing and neutral mutations
num_true = len(destable)
num_false = len(neutral)

# Create figure and iterate over models again
fig,ax = plt.subplots(2,3,figsize=(10,10),dpi=300)

for i in range(len(models)):
    # Model of interest
    model = "esm1v_t33_650M_UR90S_"+str(i+1)
    
    # Create dataframes ordered by metric of interest
    t4_pse_difference_ordered = t4_df.sort_values(["log_mut_minus_wt_vals_"+model])
    t4_pse_ordered = t4_df.sort_values(["logP_mut_"+model])

    # Iterate over rows of data frame ordered by log_mut_minus_wt_vals, create data points for ROC curve:
    roc_mt_minus_wt_x = [0]
    roc_mt_minus_wt_y = [0]
    area_mt_min_wt = 0
    for j,row in t4_pse_difference_ordered.iterrows():
        if row['effect'] == 'Destabilizing':
            roc_mt_minus_wt_y.append(roc_mt_minus_wt_y[-1]+(1/num_true))
            roc_mt_minus_wt_x.append(roc_mt_minus_wt_x[-1])
        else:
            roc_mt_minus_wt_y.append(roc_mt_minus_wt_y[-1])
            roc_mt_minus_wt_x.append(roc_mt_minus_wt_x[-1]+(1/num_false))
            area_mt_min_wt += (1/num_false)*roc_mt_minus_wt_y[-1] # calculate the area under the curve

    # Same, but for data frame order by logP_mut        
    roc_mt_x = [0]
    roc_mt_y = [0]
    area_mt = 0
    for k,row in t4_pse_ordered.iterrows():
        if row['effect'] == 'Destabilizing':
            roc_mt_y.append(roc_mt_y[-1]+(1/num_true))
            roc_mt_x.append(roc_mt_x[-1])
        else:
            roc_mt_y.append(roc_mt_y[-1])
            roc_mt_x.append(roc_mt_x[-1]+(1/num_false))
            area_mt += (1/num_false)*roc_mt_y[-1]

    # Make Roc curve
    ax.flat[i].axis(xmin = -0.1, xmax =1.1, ymin = -0.1, ymax = 1.1)
    ax.flat[i].plot(roc_mt_minus_wt_x,roc_mt_minus_wt_y, label = "$\logP_{mut}-\logP_{wt}$")
    ax.flat[i].plot(roc_mt_x,roc_mt_y, label = "$\logP_{mut}$")

    # Make text box for area under the curve
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    textstr_mt_minus_wt = '\n'.join((
        r'Area under $mut-wt$',
        str(round(area_mt_min_wt,4)),
        ))

    textstr_mt = '\n'.join((
        r'Area under $mut$',
        str(round(area_mt,4)),
        ))

    # Add text box and labels,
    ax.flat[i].set_title("ROC Curve esm "+str(i+1))
    # plt.xlabel("Specificity")
    # plt.ylabel("Sensitivity")
    ax.flat[i].legend()
    # ax.flat[i].text(1, 0.5, textstr_mt_minus_wt, fontsize=8, transform=ax.flat[i].gcf().transFigure,  verticalalignment='top', bbox=props)
    # ax.flat[i].text(1, 0.25, textstr_mt, fontsize=8, transform=ax.flat[i].gcf().transFigure,  verticalalignment='top', bbox=props)
    ax.flat[i].text(.4, 0.5, textstr_mt_minus_wt, fontsize=8, verticalalignment='top', bbox=props)
    ax.flat[i].text(.4, 0.25, textstr_mt, fontsize=8, verticalalignment='top', bbox=props)
    
    # # Sanity check
    # print("Area under logP_mut-logP_wt curve for esm "+str(i+1)+":", area_mt_min_wt)
    # print("Area under logP_mutcurve for esm "+str(i+1)+":", area_mt)
    # print("\n")

ax.flat[5].axis("off")
plt.savefig("ROC_curves_full_all_esm.png",format="png") # Save the plot: 
plt.show()
```
