```python
def vector_correlation(vecs_data,vecs_pred,beta=1,inf_vals=None):
    # turn nans and non folding data into -inf
    if inf_vals != None:
        vecs_data[np.logical_and.reduce(
            [vecs_data == x for x in inf_vals]
        )] = -1 * np.inf
    
    avg_pred_prob = np.mean(softmax(vecs_pred * beta,axis=-1),axis=0)
    pred_prob_delta = softmax(vecs_pred*beta,axis=-1) - avg_pred_prob[None,:]

    avg_data_prob = np.mean(softmax(vecs_data,axis=-1),axis=0)
    data_prob_delta = softmax(vecs_data,axis=-1) - avg_data_prob[None,:]
    pred_dot_data = np.einsum(
        'na,na->n',
        pred_prob_delta,
        data_prob_delta)
    pred_dot_pred = np.einsum(
        'na,na->n',
        pred_prob_delta,
        pred_prob_delta)
    data_dot_data = np.einsum(
        'na,na->n',
        data_prob_delta,
        data_prob_delta)
    vec_corr = np.mean(pred_dot_data) / (
        np.sqrt(np.mean(pred_dot_pred) * np.mean(data_dot_data))
    )
    vec_corr_sites = pred_dot_data / (
        np.sqrt(pred_dot_pred * data_dot_data)
    )
    return vec_corr,vec_corr_sites
```

```python
data_matrix = np.random.normal(size=(100,20))
pred_matrix = np.random.normal(size=(100,20))
test_labels = np.random.randint(low=0,high=20,size=(100))
summary,sites = vector_correlation(
            data_matrix,
            pred_matrix,
            beta=0.15,
            inf_vals=[-4]
        )

newmap = keep_center_colormap(
        np.nanmin(data_matrix[data_matrix != -np.inf]),
        np.nanmax(data_matrix[data_matrix != -np.inf]),
    )

fig,ax = plt.subplots(1,3,figsize=(20,100),dpi=50,gridspec_kw={'width_ratios': [20,20,1]})
im = ax[0].imshow(
    data_matrix,
    cmap=newmap,
              )
#ax.colorbar()
ax[0].set_yticks(range(len(data_matrix)))
ax[0].set_yticklabels(labels=['{:>2} {:>3}'.format(y,x) 
                              for x,y in zip(np.arange(len(data_matrix)),
                                             [aa_to_one_letter[ind_to_aa_size[x]] for x in test_labels])])
ax[0].set_xticks(range(18))
ax[0].set_xticklabels(labels=[ind_to_ol_nisthal[x] for x in range(18)])
fig.colorbar(im,ax=ax[0])
for i,idx in enumerate(test_labels):
    wt_idx = ol_to_ind_nisthal[aa_to_one_letter[ind_to_aa_size[idx]]]
    #print(i)
    x_cell(wt_idx,i,ax=ax[0],color='black',marker='.',s=8,linewidth=1)
    mut_color = 'black'

ax[0].set_title('Stability')


newmap = keep_center_colormap(
    np.min(pred_matrix),
    np.max(pred_matrix))    
im1 = ax[1].imshow(pred_matrix,
               cmap=newmap,
              )
#ax.colorbar()
ax[1].set_yticks(range(len(data_matrix)))
ax[1].set_yticklabels(labels=[
    '{:>2} {:>3}'.format(y,x) for x,y in zip(np.arange(len(data_matrix)),
                                             [aa_to_one_letter[ind_to_aa_size[x]] for x in test_labels])])
ax[1].set_xticks(range(20))
ax[1].set_xticklabels(labels=[ind_to_ol_nisthal[x] for x in range(20)])
fig.colorbar(im1,ax=ax[1])
for i,idx in enumerate(test_labels):
    wt_idx = ol_to_ind_nisthal[aa_to_one_letter[ind_to_aa_size[idx]]]
    #print(i)
    x_cell(wt_idx,i,ax=ax[1],color='black',marker='.',s=8,linewidth=1)
    mut_color = 'black'
ax[1].set_title(r'$\log P^{\alpha} - \log P^{wt} (\beta = 0.15)$')    



corr_map = keep_center_colormap(np.min(sites),1.1,cmap='PiYG')
im2 = ax[2].imshow(
    sites[:,None],
    cmap=corr_map
)
fig.colorbar(im2,ax=ax[2])

for nan_coords in np.argwhere(np.isinf(data_matrix)):
    #print(i)
    x_cell(
           nan_coords[1],
        nan_coords[0],
        marker='x',
           ax=ax[0],color='black',s=100,linewidth=1)
    x_cell(
           0,
        nan_coords[0],
        marker='x',
           ax=ax[2],color='black',s=100,linewidth=1)
    mut_color = 'black'
    
ax[2].set_title(r'Vector correlation')    
ax[2].set_yticks(range(len(data_matrix)))
ax[2].set_yticklabels(labels=['{:>3}'.format(x) for x in (np.arange(len(data_matrix)))])
ax[2].set_xticks([])
fig.tight_layout()
# plt.savefig('ProteinG_data_predictions_vec_corr.pdf')
plt.show();
```
